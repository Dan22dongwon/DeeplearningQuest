## 텐서(Tensor)란?
텐서(Tensor)는 데이터를 담기위한 컨테이너(container)로서 다차원 배열 또는 리스트 형태와 유사합니다. 일반적으로 수치형 데이터를 저장하고, 동적 크기를 가집니다. 텐서플로우(TensorFlow)는 데이터 표현과 다양한 수학식을 계산하기 위한 기본 구조로 텐서를 사용해서 표현합니다.

텐서는 형상이 존재하며 데이터의 표현에 따라서 사용합니다. 사용되는 용어는 다음과 같습니다.

Rank: 축(차원)의 개수
Shape: 형상(각 축의 요소의 개수)
Type: 데이터 타입

**먼저 텐서를 생성하기 위해 텐서플로우(TensorFlow)를 임포트합니다.**

```python
import tensorflow as tf
```

### **0D Tensor(Scalar)**

0차원 텐서는 하나의 숫자를 담고 있는 텐서(tensor)입니다. 스칼라(scalar)라고도 부르며, 축과 형상이 없습니다.

tf.constant()를 사용하면 상수(constant) 텐서를 만들 수 있고, 만든 텐서에 tf.rank()를 적용하면 축의 개수를 알 수 있습니다.

```python
t0 = tf.constant(1)
print(t0)
print(tf.rank(t0)) # 축의 개수를 반환합니다.
```

### **1D Tensor(Vector)**

```python
t1 = tf.constant([1, 2, 3])
print(t1)
print(tf.rank(t1))
```

### **2D Tensor(Matrix)**
2차원 텐서는 행렬과 같은 모양으로 두개의 축이 존재합니다. 일반적인 수치, 통계 데이터셋이 여기에 해당됩니다. 주로 샘플(samples)과 특성(features)을 가진 구조로 사용됩니다.

```python
# Q. 위 이미지와 같은 크기 (3, 3)의 2차원 텐서를 만들고, 차원을 반환해 주세요.
# [[YOUR CODE]]
t2 = tf.constant([[1, 2, 3],[1, 2, 3],[1, 2, 3]])
print(t2)
print(tf.rank(t2))
```

### 텐서 변환
텐서 타입 변환
텐서의 타입을 변환하고자 할 때는 tf.cast를 사용합니다. 여기서는 16비트 실수형 tf.float16을 32비트 실수형 tf.float32로 변환하였습니다.

```python
f32 = tf.cast(f16, tf.float32)
print(f32)
```

다음 예제에서는 8비트 정수형 tf.int8을 32비트 정수형인 tf.int32로 변환하고자 합니다. 직접 코드를 작성해 보세요.

```python
#  Q. 8비트 정수형을 32비트 정수형으로 변환해 보세요.
# [[YOUR CODE]]
# 8비트 정수형 텐서 생성
int8_tensor = tf.constant([-1, 0, 1, 2, 3], dtype=tf.int8)

# 32비트 정수형으로 변환
int32_tensor = tf.cast(int8_tensor, tf.int32)

print(int32_tensor)
```

### 텐서 형상 변환
텐서의 형상을 변환하는 것은 tf.reshape 함수를 통해 가능합니다. 이 함수는 텐서의 원소는 그대로 유지하면서 텐서의 구조를 바꿉니다.

```python
x = tf.constant([[1], [2], [3]])
print(x)
print(x.shape)

y = tf.reshape(x, [1, 3])
print(y)
print(y.shape)
```

### 텐서 전치
텐서를 전치하여 형상을 바꾸는 역할로 tf.transpose 함수를 사용할 수 있습니다

```python
print(y)
print(tf.transpose(y))
print(y.shape)
```

### 차원 압축
텐서에서 크기가 1인 차원을 제거하는 tf.squeeze 함수를 이용해 형상을 변경합니다.

```python
print(x)
print(tf.squeeze(x))
```

### 차원 추가
텐서의 차원을 추가하는 tf.expand_dims 함수를 이용해 형상을 변경합니다. 여기서 axis는 차원을 확장할 텐서의 축을 지정해주는 역할을 합니다.

```python
print(y)
print(tf.expand_dims(y, axis=0))
print(tf.expand_dims(y, axis=1))
print(tf.expand_dims(y, axis=2))
```

### 텐서 변환 정리
위에서 사용한 함수들 외에서 텐서플로우에서 제공되는 여러 텐서 함수들이 존재합니다. 다양한 텐서 함수를 자유롭게 연습해 보세요.

image.png

```python
# [Playground]] 다양한 텐서 함수를 연습해 보세요.
# [[YOUR CODE]]

# 텐서 생성
tensor_a = tf.constant([1, 2, 3, 4, 5], dtype=tf.float32)
tensor_b = tf.constant([6, 7, 8, 9, 10], dtype=tf.float32)

# 텐서 합치기
concat_tensor = tf.concat([tensor_a, tensor_b], axis=0)
print("Concatenated Tensor:")
print(concat_tensor)

# 텐서 곱하기
multiply_tensor = tf.multiply(tensor_a, tensor_b)
print("\nMultiplied Tensor:")
print(multiply_tensor)

# 텐서의 평균 구하기
mean_tensor = tf.reduce_mean(tensor_a)
print("\nMean of Tensor A:")
print(mean_tensor)

# 텐서의 최댓값과 최솟값 구하기
max_value = tf.reduce_max(tensor_b)
min_value = tf.reduce_min(tensor_b)
print("\nMax Value of Tensor B:", max_value)
print("Min Value of Tensor B:", min_value)
```

### 0차원 텐서의 연산
텐서는 기본 산술과 다양한 연산을 수행 할 수 있습니다.

먼저 더하기와 빼기 연산을 수행해보겠습니다. 더하기 연산자인 +와 빼기 연산자인 -를 이용해서 텐서간의 계산이 가능하고, add와 subtract 함수를 이용해서도 계산이 가능합니다.

```python
print(tf.constant(2) + tf.constant(2))
print(tf.constant(2) - tf.constant(2))
print(tf.add(tf.constant(2), tf.constant(2)))
print(tf.subtract(tf.constant(2), tf.constant(2)))
```

곱하기와 나누기 연산도 *와 / 연산자를 이용하는 방법과 multiply와 divide 함수를 이용하는 방법이 있습니다.

```python
print(tf.constant(2) * tf.constant(2))
print(tf.constant(2) / tf.constant(2))
print(tf.multiply(tf.constant(2), tf.constant(2)))
print(tf.divide(tf.constant(2), tf.constant(2)))
```

텐서 연산에서 주의할 점은 서로 다른 타입을 가지는 텐서는 연산이 되지 않고, 에러가 발생한다는 점입니다.

```python
print(tf.constant(2) + tf.constant(2.2))
```

다른 타입을 가지는 텐서는 연산을 위해서 타입을 변환해 주어야합니다. 2-3 스텝에서 언급했던 tf.cast를 사용하여 텐서의 타입을 서로 맞춰준 후에 연산을 할 수 있습니다.

```python
print(tf.cast(tf.constant(2), tf.float32) + tf.constant(2.2))
```

### 1차원 이상의 텐서 연산
0차원의 상수값만 존재하는 텐서 계산 외에도 1차원 이상의 텐서에 대해서도 연산이 가능합니다. 다만 텐서의 모양이 직사각형이거나 연산이 가능하도록 형상을 맞춰주어야 합니다.

### 1차원 텐서의 연산
1차원 텐서 2개를 만들어 봅시다.

```python
# Q. 1차원 텐서 2개를 생성해 보세요.
# [[YOUR CODE]]
# 1차원 텐서 생성
tensor_a = tf.constant([1, 2, 3, 4, 5], dtype=tf.float32)
tensor_b = tf.constant([6, 7, 8, 9, 10], dtype=tf.float32)

print("Tensor A:")
print(tensor_a)

print("\nTensor B:")
print(tensor_b)

# Q. 사칙연산(더하기, 빼기, 곱하기, 나누기)을 해보세요.
# [[YOUR CODE]]

# 덧셈
add_result = tf.add(tensor_a, tensor_b)
print("Addition Result:")
print(add_result)

# 뺄셈
subtract_result = tf.subtract(tensor_a, tensor_b)
print("\nSubtraction Result:")
print(subtract_result)

# 곱셈
multiply_result = tf.multiply(tensor_a, tensor_b)
print("\nMultiplication Result:")
print(multiply_result)

# 나눗셈
divide_result = tf.divide(tensor_a, tensor_b)
print("\nDivision Result:")
print(divide_result)
```

### 2차원 텐서의 연산
2차원 텐서를 생성한 후 연산을 해 봅시다.

```python
# Q. 크기 (2,2)인 2차원 텐서 2개를 생성해 보세요.
# [[YOUR CODE]]


# 크기가 (2,2)인 2차원 텐서 생성
c = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
d = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

print("Tensor C:")
print(c)

print("\nTensor D:")
print(d)
```

텐서를 계산하기 위한 여러 연산자들을 제공하고 있으며, +, -, *, / 연산자는 요소별(element-wise) 연산을 수행하고, @ 연산자의 경우에는 행렬 곱 연산을 수행합니다.

```python
print(c + d) # element-wise addition
print(c - d) # element-wise subtraction
print(c * d) # element-wise multiplication
print(c @ d) # matrix multiplication
print(c / d) # element-wise division
```

텐서 연산은 연산자를 통해서도 가능하지만, 제공되는 함수를 통해서도 연산이 가능합니다.

```python
print(tf.add(c, d))
print(tf.subtract(c, d))
print(tf.multiply(c, d))
print(tf.matmul(c, d))
print(tf.divide(c, d))
```

그 밖에도 유용한 여러 함수들이 존재합니다.

- reduce_max(): 텐서 값 중에서 최대값을 계산합니다.
- argmax(): 최대값의 위치를 반환합니다. 예제에서는 [4.0, 5.0, 6.0] 보다 [10.0, 9.0, 8.0]이 다 크기 때문에 [0 0 0]이 아니라 [1 1 1]의 결과를 보여주고 있습니다.
- nn.softmax(): 텐서의 값을 0과 1 사이의 값으로 보여줍니다. 예제에서는 [4.0, 5.0, 6.0] 값이 [0.09003057 0.24472848 0.66524094]으로 변환되었고, [10.0, 9.0, 8.0]이 [0.66524094 0.24472848 0.09003057]로 변환되었습니다.

```python
E = tf.constant([[4.0, 5.0, 6.0],
                 [10.0, 9.0, 8.0]])

print(tf.reduce_max(E))
print(tf.argmax(E))
print(tf.nn.softmax(E))
```

### 텐서의 연산 정리

```python
# [Playground] 다양한 텐서의 연산을 연습해 보세요!
# [[YOUR CODE]]

import tensorflow as tf

# 크기가 (2,2)인 2차원 텐서 생성
tensor_a = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
tensor_b = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

# 행렬 곱셈
matmul_result = tf.matmul(tensor_a, tensor_b)
print("Matrix Multiplication Result:")
print(matmul_result)

# 전치 행렬
transpose_result = tf.transpose(tensor_a)
print("\nTranspose Result:")
print(transpose_result)

# 텐서의 합
sum_result = tf.reduce_sum(tensor_a)
print("\nSum of Tensor A:")
print(sum_result)

# 텐서의 제곱근
sqrt_result = tf.sqrt(tensor_b)
print("\nSquare Root of Tensor B:")
print(sqrt_result)
```
